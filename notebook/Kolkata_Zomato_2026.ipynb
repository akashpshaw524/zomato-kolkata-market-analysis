## 3. Area Analysis — Where are restaurants concentrated?
*Using Pandas value_counts, groupby, sort_values*

```python
# Q1 — Top 12 areas by number of restaurants

top_locations = df['address'].value_counts().head(12).reset_index()
print(top_locations)
top_locations.columns = ['area', 'restaurant_count']
top_locations['percentage'] = (top_locations['restaurant_count'] / len(df) * 100).round(1)
print(top_locations.to_string(index=False))
print('\nTop area:', top_locations['area'].iloc[0],
      'with', top_locations['restaurant_count'].iloc[0], 'restaurants')
```

```python
# Q2 — Average rating by area (Pandas groupby + agg)

# Step 1: Clean columns to numbers (Crucial or mean() will fail)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['cost_for_two'] = pd.to_numeric(df['cost_for_two'], errors='coerce')

# Step 2: Group and Aggregate
# We use a dictionary to tell Pandas exactly what to do with each column
area_stats = df.groupby('address').agg(
    total_restaurants = ('name', 'count'),
    avg_rating        = ('rating', 'mean'),
    avg_cost          = ('cost_for_two', 'mean'),
    total_votes       = ('voteCount', 'sum')
).reset_index()

print(area_stats)

print(area_stats.info())

# Step 3: Clean the results (Rounding)
area_stats['avg_rating'] = area_stats['avg_rating'].round(2)
area_stats['avg_cost']   = area_stats['avg_cost'].fillna(0).astype(int)

print(area_stats)

print(area_stats.info())


# Step 4: Filter and Sort
area_stats = area_stats[area_stats['total_restaurants'] >=3]
area_stats = area_stats.sort_values('avg_rating', ascending=False)

# Step 5: Display
display(area_stats.head(20))
```

## 4. Cuisine Analysis — What does Kolkata love to eat?
*Using Pandas str.split, explode, value_counts*

```python
# Q3 — Top cuisines
# Each restaurant may list multiple cuisines: 'North Indian, Chinese, Biryani'
# explode splits each into its own row — powerful Pandas technique

cusine_series = (
    df['cusine']
    .dropna()
    .str.split(',')
    .explode()
    .str.strip()
)

top_cusines = cusine_series.value_counts().head(20).reset_index()
top_cusines.columns = ['cusine', 'restaurant_count']
top_cusines['percentage'] = (top_cusines['restaurant_count']
                               / len(df) * 100).round(1)

print('=== TOP 15 CUISINES IN KOLKATA ===')
print(top_cusines.to_string(index=False))
print('\nMost popular cuisine:', top_cusines['cusine'].iloc[0])
```

```python
# Q4 Unique Restaurants

# 1. Count how many UNIQUE restaurants you have in total
total_unique_restaurants = df['name'].nunique() 

# 2. Get your top 15 cuisines
top_cuisines = cusine_series.value_counts().head(15).reset_index()
top_cuisines.columns = ['cuisine', 'restaurant_count']

# 3. Calculate percentage based on the UNIQUE total
top_cuisines['percentage'] = (top_cuisines['restaurant_count'] / total_unique_restaurants * 100).round(1)

print(f'Total Unique Restaurants in Data: {total_unique_restaurants}')
print(top_cuisines.to_string(index=False))
```

## 5. Cost Analysis — What does dining in Kolkata cost?
*Using Pandas cut, qcut for binning + groupby*

```python
# Q5 — Cost distribution using Pandas cut (binning)

bins = [0, 300, 800, 1500, 3000, 5000, 99999]
labels = ['Under 300', '301-800', '801-1500', 
          '1501-3000', '3001-5000', 'Above 5000']

df['cost_category'] = pd.cut(df['cost_for_two'],
                             bins=bins,
                             labels=labels)

# Use value_counts() WITHOUT sort - it auto-sorts by count
cost_dist = df['cost_category'].value_counts().reset_index()
cost_dist.columns = ['cost_range', 'restaurants']
cost_dist['percentage'] = (cost_dist['restaurants'] / len(df) * 100).round(1)

# Sort by the category ORDER (not by count)
cost_dist['cost_range'] = pd.Categorical(cost_dist['cost_range'], 
                                         categories=labels, 
                                         ordered=True)
cost_dist = cost_dist.sort_values('cost_range')

print("\n=== COST DISTRIBUTION (for two people) ===")
print(cost_dist.to_string(index=False))

```

```python
# Q6 — Cost vs Rating: does spending more mean better food?
# A. Ensure columns are definitely numbers
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
# 1. Remove commas and spaces just in case
df['voteCount'] = df['voteCount'].astype(str).str.replace('votes', '').str.strip()


# 2. Force to numbers (turns "garbage" text into empty NaN)
df['voteCount'] = pd.to_numeric(df['voteCount'], errors='coerce')

# 3. Fill true missing values with 0 so the sum/mean works
df['voteCount'] = df['voteCount'].fillna(0)

# Now check the data types - voteCount should be float64 or int64
print(df['voteCount'].dtype)

# B. Run the Groupby
cost_rating = (
    df.groupby('cost_category', observed=True)
    .agg(
        restaurants = ('name', 'count'),
        avg_rating  = ('rating', 'mean'),
        avg_votes   = ('voteCount', 'mean')
    )
    .reset_index()
)

# C. Clean up the result (Handle NaNs before converting to int)
cost_rating['avg_rating'] = cost_rating['avg_rating'].round(2)
cost_rating['avg_votes']  = cost_rating['avg_votes'].fillna(0).astype(int)

print('=== DOES HIGHER COST MEAN BETTER RATING? ===')
print(cost_rating.to_string(index=False))
```

## 7. Votes Analysis — Which restaurants are most popular?
*Using Pandas sort_values, nlargest, rank*

```python
# Q7 — Top 15 most voted restaurants in Kolkata
top_voted = (
    df[['name', 'address', 'rating', 'voteCount', 'cost_for_two', 'cusine']]
    .nlargest(15, 'voteCount')
    .reset_index(drop=True)
)
top_voted.index = top_voted.index + 1   # rank starts from 1

print('=== TOP 15 MOST POPULAR RESTAURANTS (by votes) ===')
print(top_voted[['name', 'address', 'rating',
                  'voteCount', 'cost_for_two']].to_string())
```

```python
# Q8 — Hidden gems: High rating but low votes
# These are underrated restaurants worth visiting
df['vote_rank']   = df['voteCount'].rank(pct=True, method='dense').round(2)
df['rating_rank'] = df['rating'].rank(pct=True, method='dense').round(2)

hidden_gems = df[
    (df['rating_rank'] >= 0.80) &   # top 20% by rating
    (df['vote_rank']   <= 0.30)      # bottom 30% by votes
][['name', 'address', 'rating', 'voteCount', 'cost_for_two', 'cusine']]

hidden_gems = hidden_gems.sort_values('rating', ascending=False).head(150)
hidden_gems = hidden_gems.reset_index(drop=True)
hidden_gems.index = hidden_gems.index + 1
print(f"Total Hidden Gems found: {len(hidden_gems)}")
print('=== HIDDEN GEMS — High Rated, Low Votes ===')
print(hidden_gems[['name', 'address', 'rating',
                    'voteCount', 'cost_for_two',]].to_string())
```

```python
# Q9 - Does popularity affect quality?
 
# Create popularity tiers based on vote_count
df_votes = df[df['voteCount'] > 0].copy()
 
df_votes['popularity_tier'] = pd.cut(
    df_votes['voteCount'],
    bins=[0, 100, 500, 2000, 100000],
    labels=['Low Votes (0-100)', 'Medium (100-500)', 'High (500-2000)', 'Very Popular (2000+)'],
    include_lowest=True
)
 
quality_by_popularity = (
    df_votes[df_votes['rating'].notna()].groupby('popularity_tier', observed=True)
    .agg(
        restaurants=('name', 'count'),
        avg_rating=('rating', 'mean'),
        min_rating=('rating', 'min'),
        max_rating=('rating', 'max')
    )
    .reset_index()
)
 
quality_by_popularity['avg_rating'] = quality_by_popularity['avg_rating'].round(2)
 
print(quality_by_popularity.to_string(index=False))
print("\nInsight: If avg_rating is similar → quality doesn't drop with popularity!")
print("\n")
```

```python
# Q10: Best Value Restaurants (Rating per rupee)

value_data = df[
    (df['rating'] >= 3.8) & 
    (df['cost_for_two'].notna()) & 
    (df['cost_for_two'] > 0) &
    (df['voteCount'] > 10)
].copy()
 
# Calculate value score (rating per ₹100)
value_data['value_score'] = value_data['rating'] / (value_data['cost_for_two'] / 100)
 
best_value = value_data.nlargest(55, 'value_score')[
    ['name', 'rating', 'cost_for_two', 'value_score', 'voteCount']
].copy()
 
best_value['cost_for_two'] = best_value['cost_for_two'].astype(int)
best_value['voteCount'] = best_value['voteCount'].astype(int)
best_value['value_score'] = best_value['value_score'].round(2)
print(f"Total Best Value Restaurants found: {len(best_value)}")
print(best_value.to_string(index=False))
print("\nInsight: These restaurants offer excellent quality at reasonable prices!")
print("\n")
 
```

## 8. NumPy Deep Statistical Analysis
*Pure NumPy operations — mean, median, std, percentile*

```python
# NumPy arrays from cleaned data
ratings_arr = np.array(df['rating'].dropna())
cost_arr    = np.array(df['cost_for_two'].dropna())
votes_arr   = np.array(df['voteCount'].dropna())

print('=' * 50)
print('  NUMPY — RATING STATISTICS')
print('=' * 50)
print('Total restaurants :', len(ratings_arr))
print('Mean rating       :', round(np.mean(ratings_arr), 2))
print('Median rating     :', round(np.median(ratings_arr), 2))
print('Std deviation     :', round(np.std(ratings_arr), 2))
print('Variance          :', round(np.var(ratings_arr), 2))
print('Min rating        :', round(np.min(ratings_arr), 1))
print('Max rating        :', round(np.max(ratings_arr), 1))
print('25th percentile   :', np.percentile(ratings_arr, 25))
print('50th percentile   :', np.percentile(ratings_arr, 50))
print('75th percentile   :', np.percentile(ratings_arr, 75))
print('IQR               :', round(
    np.percentile(ratings_arr, 75) - np.percentile(ratings_arr, 25), 2))
```

```python
print('=' * 50)
print('  NUMPY — COST FOR TWO STATISTICS')
print('=' * 50)
print('Mean cost         : Rs.', round(np.mean(cost_arr)))
print('Median cost       : Rs.', round(np.median(cost_arr)))
print('Std deviation     : Rs.', round(np.std(cost_arr)))
print('Min cost          : Rs.', int(np.min(cost_arr)))
print('Max cost          : Rs.', int(np.max(cost_arr)))
print('25th percentile   : Rs.', np.percentile(cost_arr, 25))
print('75th percentile   : Rs.', np.percentile(cost_arr, 75))
print('90th percentile   : Rs.', np.percentile(cost_arr, 90))

print('\n' + '=' * 50)
print('  NUMPY — VOTES STATISTICS')
print('=' * 50)
print('Mean votes        :', round(np.mean(votes_arr)))
print('Median votes      :', round(np.median(votes_arr)))
print('Max votes         :', int(np.max(votes_arr)))
print('Std deviation     :', round(np.std(votes_arr)))
```

```python
# NumPy where — classify restaurants as budget, mid, premium
cost_arr_full = np.array(df['cost_for_two'])

segment = np.where(cost_arr_full <= 300,   'Budget',
          np.where(cost_arr_full <= 800,   'Mid-Range',
          np.where(cost_arr_full <= 1500,  'Premium',
                                           'Luxury')))

df['segment'] = segment

unique, counts = np.unique(segment, return_counts=True)

print('=' * 45)
print('  NUMPY WHERE — RESTAURANT SEGMENTS')
print('=' * 45)
for seg, cnt in zip(unique, counts):
    pct = round(cnt / len(segment) * 100, 1)
    bar = '█' * int(pct / 2)
    print(seg.ljust(12), str(cnt).rjust(5),
          'restaurants', str(pct).rjust(5), '%', bar)
```

```python
# Customer engagement pattern

print("\n" + "="*50)
print("  NUMPY — CUSTOMER ENGAGEMENT PATTERNS")
print("="*50)

votes_arr = np.array(df['voteCount'].dropna())

# Engagement Tiers
super_popular = np.sum(votes_arr > 2000)
niche = np.sum(votes_arr < 100)
average = np.sum((votes_arr >= 100) & (votes_arr <= 2000))

print(f"\n🔥 Super Popular (>2,000 votes): {super_popular} restaurants")
print(f"📍 Niche (<100 votes):          {niche} restaurants")
print(f"📊 Average (100-2,000 votes):   {average} restaurants")

# Percentage breakdown
total_rated = len(votes_arr)
print(f"\nBreakdown (out of {total_rated} restaurants with votes):")
print(f"  Super Popular: {(super_popular/total_rated*100):.1f}%")
print(f"  Niche:        {(niche/total_rated*100):.1f}%")
print(f"  Average:      {(average/total_rated*100):.1f}%")

print(f"\nInsight: Top {super_popular} restaurants get majority of votes!")
```
